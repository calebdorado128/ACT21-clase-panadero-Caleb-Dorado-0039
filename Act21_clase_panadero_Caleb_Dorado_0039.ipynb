## Nombre de la clase: Panadero
![](https://raw.githubusercontent.com/calebdorado128/ACT21-clase-panadero-Caleb-Dorado-0039/refs/heads/main/infoclasepanadero2.png)

```python
# Definición de la clase
class Panadero:
    # Atributos de la clase (Constructor)
    def __init__(self, nombre, telefono, edad):
        self.nombre = nombre
        self.telefono = telefono
        self.edad = edad

    # Función 1: Preparar la masa
    def prepararlamasa(self):
        print(f"{self.nombre} está reuniendo la harina y el agua para la mezcla.")

    # Función 2: Amasar
    def amasar(self):
        print(f"{self.nombre} está amasando con fuerza. ¡La masa está quedando elástica!")

    # Función 3: Hornear
    def hornear(self):
        print(f"El horno está listo. {self.nombre} ha metido el pan a 180°C.")

# --- Creación de Objetos (Instancias) ---

# Objeto 1
panadero1 = Panadero("Juan Pérez", "555-0123", 35)

# Objeto 2
panadero2 = Panadero("María García", "555-0456", 28)

# --- Uso de los objetos y sus funciones ---

print(f"--- Datos del Panadero 1 ---")
print(f"Nombre: {panadero1.nombre}")
print(f"Teléfono: {panadero1.telefono}")
print(f"Edad: {panadero1.edad}")
panadero1.prepararlamasa()
panadero1.amasar()
panadero1.hornear()

print("\n" + "-"*30 + "\n")

print(f"--- Datos del Panadero 2 ---")
print(f"Nombre: {panadero2.nombre}")
print(f"Teléfono: {panadero2.telefono}")
print(f"Edad: {panadero2.edad}")
panadero2.prepararlamasa()
panadero2.amasar()
panadero2.hornear()

print("\n" + "-"*30 + "\n")

print("Programa elaborado por Caleb Dorado, NL = 14, NC = 0039")
```
